# Project Subject 2: Implementation of the FastFD Algorithm

```python
def FastFD(path):
    R = loadFile(path)[1]
    r = loadFile(path)[0]
    r = generateIDsForTuples(r)
    DI = generateEvidenceSets(R, r) 
    SCs = dict() # A dictionary for storing set covers for each attribute A in R
    for A in R:
        SCs[A] = setCover(A,DI)
    return SCs

def loadFile(path):
    file=open(path,'r', encoding='utf-8-sig')
    r=set()
    lines=file.readlines()
    for line in lines:
        row=tuple(line.strip().split(','))
        #print(row)
        #row.discard('')
        r.add(row)
    
    R=[]
    nb_attr=len(tuple(lines[0].strip().split(',')))
    ch ='A'
    if (nb_attr == 0):
        print("Le fichier est vide !")
    else:
        R.append(ch)
        for i in range(1,nb_attr):
            attr = chr(ord(ch) + i)
            #print(attr)
            R.append(attr)
            
    return r, R

def generateIDsForTuples(r):
    # e is a set of disciotanries. For each dictionary t in r, this method added an element (id,value), where the 
    # id is an integer that allows the unique identification of the tuples among the tuples in r.
    r1=[]
    i=0 # ceci est notre indentifiant de départ
    for t in r:
        d = dict() #On crée un nouveau dictionnaire contenant l'ancien tuple auquel on ajoute son identifiant
        d[i]=t
        r1.append(d)
        i+=1
        
    return r1



def identifyDifferentSet(ti,tj,R):
    diffset = set({})
    tiV= list(ti.values())[0]
    tjV = list(tj.values())[0]
    # return a singleton set, the element of which is a set containing the attributes that have different value in ti and tj
    for k in range(len(R)):
        if (tiV[k]!=tjV[k]):
            diffset.add(R[k])
    return diffset  

def generateEvidenceSets(R, r):
    EVs = set() # It is a set of sets
    for ti in r:
        for tj in r:
            if not({k: ti[k] for k in ti if k in tj and ti[k] == tj[k]}):
                    EVs.add(frozenset(identifyDifferentSet(ti,tj,R)))
    return EVs

  
#Construction d'un arbre en DFS et recuperation de toutes les couvertures de l'arbre
def constructTree(candidates, DIA):

    #Fonction recursive de parcours de l'arbre
    ##'combination' est un ensemble d'attributs visites dans la branche 
    ##'candidates' est un ensemble d'attributs non visites dans la branche
    def rec_tree(combination, candidates,debug):
        
        #Liste contenant DIA mis a jour a chaque noeud
        remaning=DIA.copy()
        
        for elem in DIA:
            
            current_attr=combination.split(",")
            
            #Recherche si au moins 1 attribut est present dans tous les elements de DIA
            intersect=False
            for attr in current_attr: 
                if attr in set(elem):
                    intersect=True
            
            #Si c'est vrai et qu'il reste des elements dans DIA, on ajoute l'attribut
            if (intersect): 
                remaning.remove(elem)
                if not remaning:
                    cover.append(current_attr)
            
        if candidates: #Si candidates est non vide
            
            for i in range(len(candidates)):
                #Ajout du prochain candidat non visite de la branche a la combinaison
                next_combination = combination +","+ candidates[i]
                #Appel recursif avec la nouvelle combinaison
                rec_tree(next_combination, candidates[i + 1:],debug+"   ")
            
    #Transformation du 'set' en 'list' et tri par ordre alphabetique des attributs
    candidates=list(candidates)
    candidates.sort()
    #Liste representant l'arbre en DFS
    cover = []
    
    for i in range(len(candidates)):
        #Parcours de l'arbre pour le candidat actuel
        rec_tree(candidates[i], candidates[i+1:],"")

    return cover


#Recuperation des elements d'une liste de longueur minimale
def minsize(list):
    
    min = None
    min_list=[]
    
    for elem in list:    
        if min==None: 
            min = elem        
        else:
            if len(elem) < len(min):
                min=elem
                
    for elem in list:
        if len(elem)==len(min):
            min_list.append(elem)

    return min_list

# Recherche des couvertures minimales d'un attribut A
## 'SCA' est un set contenant les couvertures minimales de A
## 'DIA' est une liste contenant les ensembles des differences modulo A
## 'candidates' est un set contenant les attributs eligibles à la couverture minimale pour A

def setCover(A,EV):
    
    SCA = set()
    DIA = []
    candidates = set() 
    
    for elem in EV:
        #Creation d'un subset pour chaque element de EV
        subsetEV=set(elem)
        
        #Selection des elements contenant A    
        if A in subsetEV :
            subsetEV.remove(A)              #On retire A des subsets
            DIA.append(list(subsetEV))      #On ajoute les subsets a DIA
            
            #Selection des candidats
            for attr in subsetEV :
                    candidates.add(attr)
    
    #print("Candidates:",candidates)
    #print("DIA:", DIA)
    #print("Tree:")
    cover = constructTree(candidates,DIA)
    min_cover = minsize(cover)
    print("Couvertures minimales pour "+A+" = ", min_cover)
    
    return cover  



#FastFD("iris.csv")
#FastFD("test.csv")
FastFD("adult.csv")
```
